### Dropping unnecessary columns

```python
df.drop(['Indicator Name','Country Name'], axis=1, inplace=True)
```

### Extraction  of the top 10 countries with respect to total population

```python
# Filter data for total population
total_population_data=df[df['Indicator Code']=='SP.POP.TOTL']

# Sort data based on the total population
total_population_sorted = total_population_data.sort_values(by="2024",ascending = False)

# Get the top ten countries with the highest total population for 2024
total_top_ten_countries = total_population_sorted.head(10)

print("Top ten countries of total population \n")
print(total_top_ten_countries[['Country Code']])
```

### Bar plot of this data

#### Top ten countries of total population in the year 2020 and 2024

```python
# Create a bar plot
plt.figure(figsize=(15,6))
plt.subplot(2,2,1)
sns.barplot(x="2024", y = "Country Code",hue="Country Code",data = total_top_ten_countries, palette = "coolwarm")
plt.title("Top ten countries of total population(2024)", fontsize=10)
plt.xlabel("Total Population",fontsize=10)
plt.ylabel("Countries", fontsize=10)
plt.show()


# Create a bar plot
plt.figure(figsize=(15,6))
plt.subplot(2,2,2)
sns.barplot(x="2020", y = "Country Code",hue="Country Code",data = total_top_ten_countries, palette = "coolwarm")
plt.title("Top ten countries of total population(2020)", fontsize=10)
plt.xlabel("Total Population",fontsize=10)
plt.ylabel("Countries", fontsize=10)
plt.show()
```

### Extraction of the top 10 countries with respect to total population from the bottom

```python
# Sort data based on the total population
total_population_sorted_1 = total_population_data.sort_values(by="2024",ascending = True)

# Get the top ten countries with the highest total population for 2024
total_bottom_ten_countries = total_population_sorted_1.head(10)

print("Bottom ten countries of total population \n")
print(total_bottom_ten_countries[['Country Code']])
```

### Bar plot of this data

### Extraction of up to 10 countries with the highest male population

```python
# Filter data for the male population 
male_population_data = df[df['Indicator Code']=='SP.POP.TOTL.MA.IN']

# Sort data based on the male population for 2024
male_population_sorted = male_population_data.sort_values(by = "2024", ascending=False)

# Get the top 10 countries with the highest male population for 2024
male_top_ten_countries = male_population_sorted.head(10)

print("Top 10 countries of male population")
print(male_top_ten_countries[['Country Code']])
```

### Extraction of up to 10 countries with the highest male population

```python
# Filter data for the male population 
female_population_data = df[df['Indicator Code']=='SP.POP.TOTL.FE.IN']

# Sort data based on the male population for 2024
female_population_sorted = female_population_data.sort_values(by = "2024", ascending=False)

# Get the top 10 countries with the highest male population for 2024
female_top_ten_countries = female_population_sorted.head(10)

print("Top 10 countries of female population")
print(female_top_ten_countries[['Country Code']])
```

### Top 10 countries with the highest male and female populations in 2024 

```python
# Create a bar plot
plt.figure(figsize=(15,6))
plt.subplot(2,2,1)
sns.barplot(x="2024", y = "Country Code",hue="Country Code",data = male_top_ten_countries, palette = "viridis")
plt.title("Top ten countries of male population(2024)", fontsize=10)
plt.xlabel("Male  Population",fontsize=10)
plt.ylabel("Countries", fontsize=10)
plt.show()


# Create a bar plot
plt.figure(figsize=(15,6))
plt.subplot(2,2,2)
sns.barplot(x="2024", y = "Country Code",hue="Country Code",data = female_top_ten_countries, palette = "coolwarm")
plt.title("Top ten countries of female population(2024)", fontsize=10)
plt.xlabel("Female Population",fontsize=10)
plt.ylabel("Countries", fontsize=10)
plt.show()
```

### Stacked Bar plot

#### Top 10 countries with male and female population

```python
# Merge male and female population data on 'Country Code'
merge_data= pd.merge(male_population_data,female_population_data, on="Country Code", suffixes=("_male","_female"))
```

```python
# Merged data
# Calculate the total population for each country (Male+Female)

merge_data["Total population"] = merge_data["2024_male"]+merge_data["2024_female"]
```

```python
merge_data.head()
```

```python
# sort data based on total population in descending order
sorted_data = merge_data.sort_values(by="Total population", ascending=False)
```

### Select the top 10 countries with the highest total population

```python
top_10_countries = sorted_data.head(10)
```

```python
# Create the stacked bar plot
plt.figure(figsize=(12,6))
sns.barplot(x="Country Code", y="2024_female",data=top_10_countries, color="red", label="female Population")
sns.barplot(x="Country Code", y="2024_male",data=top_10_countries,bottom=top_10_countries["2024_female"], color="blue", label="male Population")
plt.xlabel("country")
plt.ylabel("Population")
plt.legend()
plt.xticks(rotation=45,ha="right")
plt.show()
```

### Bottom 10 countries with male and female population(2024)

```python
# select the bottom 10 countries with the highest total population
bottom_10_countries = sorted_data.tail(10)
```

```python
# Create the stacked bar plot
plt.figure(figsize=(12,6))
sns.barplot(x="Country Code", y="2024_female",data=top_10_countries, color="red", label="female Population")
sns.barplot(x="Country Code", y="2024_male",data=top_10_countries,bottom=top_10_countries["2024_female"], color="blue", label="male Population")
plt.xlabel("country")
plt.ylabel("Population")
plt.legend()
plt.xticks(rotation=45,ha="right")
plt.show()
```
